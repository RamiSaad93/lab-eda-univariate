### Part 2: Delving into Product Pricing

**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `price` of products.
    - What's the average price point of products listed? How does this compare with the most common price point (mode)?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, range, and interquartile range for product `price`.
    - How varied are the product prices? Are there any indicators of a significant spread in prices?

3. **Visualizations**:
    - Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. *If its hard to read these diagrams, think why this is, and explain how it could be solved.*.
    - Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing. 

```python
df['price'].agg(['mean', 'median'])
```

```python
df['price'].mode()[0]
```

```python
df['price'].agg(['mean','median',lambda x: mode(x)]).rename(index={'<lambda>': 'mode'})
# or df['price'].agg(['mean','median',lambda x: x.mode()[0]]).rename(index={'<lambda>': 'mode'})
```

```python
df['price'].describe().round(2)
```

```python
variance = df['price'].var()
std = df['price'].std()
range_ = df['price'].max()-df['price'].min()
quartiles = df['price'].quantile([0.25,0.50,0.75])

print(f"The variance equales: {variance}")
print(f"The standard deviation equales: {std}")
print(f"The range equales: {range_}")
print(f"The quartiles equales: {quartiles}")
```

- Large Range, wide gap between the lowest and highest prices
- High Standard Deviation suggests that prices vary a lot from the mean or average price.
- Large Interquartile Range which indicates a significant spread in the middle 50% of prices.
- Presence of Outliers can indicate a significant spread if they are present in the dataset.

- Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. If its hard to read these diagrams, think why this is, and explain how it could be solved..
- Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.

```python
sns.histplot(df['price'], kde=True, bins=20, color="red")
```

```python
sns.boxplot(df['price'], color="red")
```

- Yes, there are products that are priced significantly higher than the rest

### Part 3: Unpacking Product Ratings

**Business Question**: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `rating` of products.
    - How do customers generally rate products? Is there a common trend?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, and interquartile range for product `rating`.
    - Are the ratings consistent, or is there a wide variation in customer feedback?

3. **Shape of the Distribution**:
    - Calculate the skewness and kurtosis for the `rating` column. 
    - Are the ratings normally distributed, or do they lean towards higher or lower values?

4. **Visualizations**:
    - Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

```python
df['stars'].agg(['mean','median',lambda x: mode(x)]).rename(index={'<lambda>': 'mode'})
```

- customers mostly rate the products by zero

```python
variance_stars = df['stars'].var()
std_stars = df['stars'].std()
range_stars = df['stars'].max()-df['stars'].min()
quartiles_stars = df['stars'].quantile([0.25,0.50,0.75])

print(f"The variance equales: {variance_stars}")
print(f"The standard deviation equales: {std_stars}")
print(f"The range equales: {range_stars}")
print(f"The quartiles equales: {quartiles_stars}")
```

```python
skewness = df['stars'].skew()
kurtosis = df['stars'].kurtosis()

print(f"The skewness equales: {skewness}")
print(f"The kurtosis equales: {kurtosis}")
```

- +ve meaning it's right skewed and the mean > mode, tail towards higher values
